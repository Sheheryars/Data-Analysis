## Q: Which year had the highest average volume of trades for Amazon and Apple?

 In this we group by the calculated average volume by the Name of the stock and by the year and use a max function to find out the highest average amongst all the years and print its year and value

```python
Amazon_df.index = pd.to_datetime(Amazon_df.index)
Amazon_df['Year'] = Amazon_df.index.year
average_volume_Amz = Amazon_df.groupby(['Year', 'Name'])['volume'].mean().rename('Avg Volume')
amazon_max_avg_volume = average_volume_Amz.loc[(slice(None), 'AMZN')].max()
amazon_max_avg_volume_date = average_volume_Amz.loc[(slice(None), 'AMZN')].idxmax()
print("Maximum average volume for Amazon (AMZN):", amazon_max_avg_volume)
print("Date corresponding to the maximum average volume:", amazon_max_avg_volume_date)

Apple_df.index = pd.to_datetime(Apple_df.index)
Apple_df['Year'] = Apple_df.index.year
average_volume_App = Apple_df.groupby(['Year', 'Name'])['volume'].mean().rename('Avg Volume')
amazon_max_avg_volume1 = average_volume_App.loc[(slice(None), 'AAPL')].max()
amazon_max_avg_volume_date1 = average_volume_App.loc[(slice(None), 'AAPL')].idxmax()
print("Maximum average volume for Amazon (AAPL):", amazon_max_avg_volume1)
print("Date corresponding to the maximum average volume:", amazon_max_avg_volume_date1)

```
