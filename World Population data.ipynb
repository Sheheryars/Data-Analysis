## Exploratory Analysis & Visualization

### Countries and their populations 

Summing the total population of each country and then plotting bar graphs to represent them 

```python
sum_of_pop = total_pop_df.sum(axis=1)
```

```python
sum_pop = sum_of_pop.sort_values(ascending = False)
```

```python
new_sum = sum_pop.head(50)
```

```python
new_index = new_sum.index
```

Making a bar graph to display top 50 countries and their respective populations 

```python
plt.figure(figsize = (17,6))
plt.xticks(rotation=75)
plt.title('Countries & their Populations')
sns.barplot(y=new_sum, x=new_index)
```

From this we can see that the majority of the population is located in China, India , United States and a few other countries, while other are exponentially lesser than those at the start

creating a pie chart to display the percentage of population distribution amongst the 10 highest populated countries

```python
used_sum == new_sum.head(10)
plt.figure(figsize=(12,6))
plt.title('Top 10 Countries population distribution')
plt.pie(used_sum, labels= used_sum.index, autopct='%1.1f%%', startangle=180);
```

We can see that China and India by far dominates the population percentage even amongst the other top 8 countries being in a 2 digit percentages unlike the other countries.

Showing the trend in population of the top 3 most populataed countries 

```python
plt.figure(figsize=(12,6))
plt.plot(total_pop_df.loc['China'], 'r-')
plt.xticks(rotation=75)
x_ticks = plt.gca().get_xticks()
plt.gca().set_xticks(x_ticks[::2])
plt.title('China')
plt.xlabel('Years')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(total_pop_df.loc['India'], 'b-')
plt.xticks(rotation=75)
x_ticks = plt.gca().get_xticks()
plt.gca().set_xticks(x_ticks[::2])
plt.title('India')
plt.xlabel('Years')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(total_pop_df.loc['United States'], 'g-')
plt.xticks(rotation=75)
x_ticks = plt.gca().get_xticks()
plt.gca().set_xticks(x_ticks[::2])
plt.title('United States')
plt.xlabel('Years')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

We can see from this data the steady growth in population through the years in all of the 3 countries. Though, it might be worth noticing that China seems to be nearing a plateau towards its end which shows that the process of population increase in China is being taken under control and their policies are affecting the over all rate of growth.

To find out which year had the most increase in population growth we will make a heatmap for the same top 3 countries

```python
total_pop_df.loc['China']
```

```python
china_pop = total_pop_df.loc['China'].diff().to_frame()
china_pop[china_pop < 0] = float('nan')
```

```python
china_pop
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(china_pop, annot=True, fmt=".1f")
plt.title('Heatmap for Population in China')
plt.ylabel('Year')
```

We can see that china's population during 1966 to 1972 and then after some fluctuations , starting from 1998 , the population increase started to decrease , being it's lowest at 2020.

```python
India_pop = total_pop_df.loc['India'].diff().to_frame()
India_pop[India_pop < 0] = float('nan')
plt.figure(figsize=(12,10))
sns.heatmap(India_pop, annot=True, fmt=".1f")
plt.title('Heatmap for Population in India')
plt.ylabel('Year')
```

We can see that there was a huge surge of growth of India's population increased during 1982 to 2006 and then started to decrease from then onwards. While we can also see that the increase in population was the least in the starting few years from 1962 to 1968.

```python
US_pop = total_pop_df.loc['United States'].diff().to_frame()
US_pop[India_pop < 0] = float('nan')
plt.figure(figsize=(12,10))
sns.heatmap(US_pop, annot=True, fmt=".1f")
plt.title('Heatmap for Population in United States')
plt.ylabel('Year')
```

We can see in this that United States did not have a very linear change in its increase in population as it fluctuated throughout the years. The highest recorded population increase is seen in year 1992 and the lowest is seen in the year of 2020.
